```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# function to plot
def a_sin(amplitude, x):
    return amplitude * np.sin(x)
```

```python
# Create the plot. 
def plot():
    x = np.linspace(-np.pi, np.pi, 201)

    fig, ax = plt.subplots(constrained_layout = True, figsize=[5,4]);

    ax.plot(x, a_sin(2., x))
    ax.set_ylabel(r'$A\sin(x)$')
    ax.set_xlabel(r'$x$')

    plt.show()
plot()
```
